```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
#read the diabetes dataset
diabetes = pd.read_csv("csv.csv")
#display this as the title
print("\nDiabetes")
#columns from the csv file as the dataframe
columns=['YEAR','STATE','RATE','DEATHS','URL']
#drop the column URL it is not needed
diabetes = diabetes.drop(columns='URL',axis=1)
#sort the columns state and year in ascending order 
diabetes = diabetes.sort_values(['STATE','YEAR'],ascending=True)
#remove all non-numeric values from the column year and replace it to NaN using errors="coerce"
diabetes["YEAR"]=pd.to_numeric(diabetes["YEAR"], errors="coerce")
#dropna removes all rows with NaN in column year
diabetes = diabetes.dropna(subset=["YEAR"])
# convert the year datatype from float to integer in order to remove the zeros
diabetes["YEAR"]=diabetes["YEAR"].astype(int)
#change the datatype of YEAR from string to float to remove the commas
diabetes["DEATHS"]= diabetes["DEATHS"].str.replace(",","").astype(float)
#convert the column deaths to float32
diabetes["DEATHS"]=diabetes["DEATHS"].astype(np.float32)
#remove all non-numeric values from the column deaths and replace it to NaN using errors="coerce"
diabetes["DEATHS"]=pd.to_numeric(diabetes["DEATHS"], errors="coerce")
#dropna removes all rows with NaN in column deaths
diabetes = diabetes.dropna(subset=["DEATHS"])
#convert the deaths datatype from string to integer in order to remove the zeros
diabetes["DEATHS"]=diabetes["DEATHS"].astype(int)
# show the diabetes dataset
display(diabetes)
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
#showing original diabetes dataset
personal_income = pd.read_csv("Personal Income.csv")
#display this as the title
print("\nPersonal Income")
#renamed column Amount to Amount Spent
personal_income.rename(columns = {'Amount': 'Amount Spent'}, inplace=True)
#renamed column Source to Payment Method
personal_income.rename(columns = {'Source': 'Payment Method'}, inplace=True)
columns = ['Posting Date', 'Day of the Week', 'Amount Spent', 'Category', 'Description', 'Payment Method']
#sort the posting date and day of the week column in order
personal_income = personal_income.sort_values(by=['Posting Date','Day of the Week'], ascending=True)
#converted column Amount spent from float to string
personal_income["Amount Spent"]= personal_income["Amount Spent"].astype(str)
#removed the - sign from the column Amount Spent
personal_income["Amount Spent"]= personal_income["Amount Spent"].str.replace("-","").astype(float)
#dropna removes all rows with NaN in column Description
personal_income = personal_income.dropna(subset=["Description"])
#replace Makeup/Skin Care in description column to Cosemetics in order to keep the data clean without / sign
personal_income["Description"].replace({"Makeup/Skin Care": "Cosmetics"}, inplace=True)
#checks in the defined column description to look for missing values and drops it
personal_income.dropna(subset=['Description'])
#show the dataset
display(personal_income)
```
